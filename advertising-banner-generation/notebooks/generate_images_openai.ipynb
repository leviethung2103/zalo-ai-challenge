## Dall-E Evaluation

```python
import csv
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
MODEL = 'dall-e-2'
IMAGE_DOWNLOAD_FOLDER = "/Users/mac/Downloads/zalo-ai-challenge/advertising-banner-generation/download"

client = OpenAI(api_key=OPENAI_API_KEY)


def download_image(url, file_path):
    response = requests.get(url)
    
    if response.status_code == 200:
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print("Image downloaded successfully!")
    else:
        print("Failed to download the image.")


def generate_image(prompt):
    image_url = None

    try:
        response = client.images.generate(
            model=MODEL,
            prompt=prompt,
            size="1024x1024",
            quality="standard",
            n=1
        )
        image_url = response.data[0].url
        print("Generated url: ", image_url)
        
    except Exception as error:
        print (f"Open AI error: {error}. Prompt: {prompt}")
        
    return image_url

def read_csv_file(filename):
    captions = []
    descriptions = []
    more_infos = []
    filenames = []

    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row

        for row in reader:
            caption = row[1]
            description = row[2]
            more_info = row[3]
            filename = row[4]

            captions.append(caption)
            descriptions.append(description)
            more_infos.append(more_info)
            filenames.append(filename)

    return captions, descriptions, more_infos, filenames


# INPUT CSV
filename = "/Users/mac/Downloads/zalo-ai-challenge/advertising-banner-generation/test/info.csv"
captions, descriptions, more_infos, filenames = read_csv_file(filename)

for i in range(len(captions)):
    save_path = os.path.join(IMAGE_DOWNLOAD_FOLDER, filenames[i])
    if os.path.exists(save_path):
        continue
    prompt = captions[i] + " " + descriptions[i] + " " + more_infos[i]
    image_url = generate_image(prompt)
    if image_url is None:
        continue

    download_image(image_url, save_path)
```
